```python
expt = "full-10"
segment = 3

dim = "y"
otherdim = "y"
if dim == "x":
    otherdim = "y"


z = xr.open_dataset(f"/g/data/nm03/ab8992/ttide-inputs/{expt}/forcing/tz_00{segment}.nc")
u = xr.open_dataset(f"/g/data/nm03/ab8992/ttide-inputs/{expt}/forcing/tu_00{segment}.nc")
mask = xr.open_dataset(f"/g/data/nm03/ab8992/ttide-inputs/{expt}/ocean_mask.nc").mask.isel({f"n{otherdim}":-1})

for i in range(len(mask) - 1):
    if mask[i] != mask[i + 1]:
        if mask[i + 1] == 1:
            land_index = i + 1
        else:
            land_index = i

        break
land_index *= 2
fig,ax = plt.subplots(1,2,figsize = (12,6))
z[f"zamp_segment_00{segment}"].plot(marker = "x",ax = ax[0])
z[f"zphase_segment_00{segment}"].plot(marker = "x",ax = ax[0])
ax[0].set_xlim(land_index - 10,land_index + 10)
ax[0].set_title("eta")
u[f"uamp_segment_00{segment}"].plot(marker = "x",ax = ax[1])
u[f"vamp_segment_00{segment}"].plot(marker = "x",ax = ax[1])
u[f"uphase_segment_00{segment}"].plot(marker = "x",ax = ax[1])
u[f"vphase_segment_00{segment}"].plot(marker = "x",ax = ax[1])
ax[1].set_xlim(land_index - 10,land_index + 10)
ax[1].set_title("Velocities")

plt.figure()

## MAKE MODIFICATIONS HERE
for i in ["phase","amp"]:
    for j in ["u","v"]:
        u[f"{j}{i}_segment_00{segment}"] = u[f"{j}{i}_segment_00{segment}"].where(
            u[f"{j}{i}_segment_00{segment}"][f"n{dim}_segment_00{segment}"] > land_index + 3,
            u[f"{j}{i}_segment_00{segment}"].isel({f"n{dim}_segment_00{segment}":land_index + 3})
            )
    j = "z"
    z[f"{j}{i}_segment_00{segment}"] = z[f"{j}{i}_segment_00{segment}"].where(
    z[f"{j}{i}_segment_00{segment}"][f"n{dim}_segment_00{segment}"] > land_index + 3,
    z[f"{j}{i}_segment_00{segment}"].isel({f"n{dim}_segment_00{segment}":land_index + 3})
    )



##

fig,ax = plt.subplots(1,2,figsize = (12,6))
z[f"zamp_segment_00{segment}"].plot(marker = "x",ax = ax[0])
z[f"zphase_segment_00{segment}"].plot(marker = "x",ax = ax[0])
ax[0].set_xlim(land_index - 10,land_index + 10)
ax[0].set_title("eta")
u[f"uamp_segment_00{segment}"].plot(marker = "x",ax = ax[1])
u[f"vamp_segment_00{segment}"].plot(marker = "x",ax = ax[1])
u[f"uphase_segment_00{segment}"].plot(marker = "x",ax = ax[1])
u[f"vphase_segment_00{segment}"].plot(marker = "x",ax = ax[1])
ax[1].set_xlim(land_index - 10,land_index + 10)
ax[1].set_title("Velocities")
print(land_index)
```
